```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.base import clone
from rfpimp import *

def mkdf(columns, importances):
    I = pd.DataFrame(data={'Feature':columns, 'Importance':importances})
    I = I.set_index('Feature')
    I = I.sort_values('Importance', ascending=False)
    return I
```

## Classification - Accuracy

```python
df = pd.read_csv("data/imp_R_class_acc.csv")
I = mkdf(df.iloc[:,0], df.iloc[:,1])
I['Importance'] = I.Importance/100.0
viz = plot_importances(I, color='#FDDB7D', title="R feature importance via drop in accuracy")
viz.save('../article/images/cls_permute_random_R.svg')
viz.save('../article/images/cls_permute_random_R.pdf')
viz
```
